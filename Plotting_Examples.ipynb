```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
#Using Jupter Notebooks as an example
#Once again, very very similar to Matlab Plotting
#Source is "Finance for Finance" -> Hilpisch
```

```python
np.random.seed(2000) #randomly generating data for plots
y = np.random.standard_normal((20,2)).cumsum(axis=0) #cumsum is called to calculate the cumlative sum of the sample data along axis 0
```

```python
plt.figure(figsize=(6,4))
plt.plot(y, lw=1.4) # Remeber that this plots two lines because we have a 20x2 matrix that was created above.
plt.plot(y, 'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Plot Example 1')
plt.show()
```

```python
fig, ax1 = plt.subplots() #plots the first data set using the first (left) axis
#plt.subplot(211) creates two seperate subplots -> Will need to comment line above if used
plt.plot(y[:,0], 'b', label='1st')
plt.plot(y[:,0], 'ro')
plt.grid(True)
plt.legend(loc=8)
plt.axis('tight')
plt.xlabel('x-axis')
plt.ylabel('Y value 1st')

#plt.subplot(212) creates 2nd subplot -> comment out the title below and ax2
plt.title('Plot Example 2')
ax2 = ax1.twinx() #plots second data set using the second (right) axis
plt.plot(y[:, 1], 'g', label = '2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('Y value 2nd')

plt.show()
```

```python
#Scatter Plot Example
#To makes things fancy, you can add a colorbar on the right side
y2 = np.random.standard_normal((1000,2))
plt.figure(figsize=(7,5))
plt.plot(y2[:,0], y2[:,1], 'ro', marker ='o')
plt.grid(True)
plt.title('Plot Example 3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
#Histogram Example
#https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html
```

```python
y3 = np.random.standard_normal((100,2))
plt.hist(y3, label=['1st data set', '2nd data set'], color=['r','b'], bins=25) #settings stacked=True is an option to stack dataset1 on dataset2
plt.grid(True)
plt.legend(loc=1)
plt.title('Plot Example 4')
plt.xlabel('x-axis')
plt.ylabel('frequency')

plt.show()
```

```python
def func(x):
    return 2*np.exp(x) + 1

a, b = 1, 1.5 # limits of integrations
x = np.linspace(0,2) #this can substited for scrapped data
y = func(x)

fig, ax = plt.subplots(figsize=(7,5))
plt.plot(x,y, 'b')
plt.ylim(ymin=0)
plt.grid(True)

Ix = np.linspace(a,b)
Iy = func(Ix) #evaluating at Ix 
verts = [(a,0)] + list(zip(Ix, Iy)) + [(b,0)]
poly = Polygon(verts, facecolor = '0.9', edgecolor = '0.3')
ax.add_patch(poly) #shaded area added

#can add the values of f(a) & f(b)

plt.show()
```
